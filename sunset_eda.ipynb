```python
import pandas as pd
import requests
from datetime import datetime
import time
import numpy as np
```

```python
sunset_qual = pd.read_csv("sunset_scores.csv")[["Date", 
                                                "City", 
                                                "Raw Number of Posts", 
                                                "Residual After Controlling for Time", 
                                                "Good Sunset"]]
sunset_qual
```

```python
sunset_qual = sunset_qual.drop(["Residual After Controlling for Time", "Raw Number of Posts"], axis = 1)
sunset_qual
```

```python
sunset_qual["Date"] = pd.to_datetime(sunset_qual["Date"])
sunset_qual["City"] = sunset_qual["City"].str.replace("_", " ")
sunset_qual
```

```python
# get lat and long of all cities in dataset
cities = sunset_qual["City"].unique()
city_lat_long = {"Los Angeles": ["34.05", "118.24"],
                "San Francisco": ["37.77", "122.41"], 
                "Chicago": ["41.87", "87.62"],
                "Miami": ["25.76", "80.19"],
                "Philadelphia": ["39.95", "75.16"],
                "NYC": ["40.71", "74.00"],
                "San Diego": ["32.71", "117.16"],
                "Boston": ["42.36", "71.05"],
                "Washington DC": ["38.90", "77.03"],
                "Seattle": ["47.60", "122.33"]}
```

```python
# list of urls for sunset times 
time_url = []
for i in range(len(sunset_qual)):
    url = "https://api.sunrise-sunset.org/json?"
    lat = "lat=" + str(city_lat_long[sunset_qual.iloc[i]["City"]][0])
    long = "&lng=" + str(city_lat_long[sunset_qual.iloc[i]["City"]][1])
    date = "&date=" + str(sunset_qual.iloc[i]["Date"].date())
    url = url + lat + long + date
    time_url.append(url)
```

```python
sunset_times = []
for url in time_url:
    time.sleep(np.random.uniform(0, 100/1000))
    utc_time = requests.get(url).json()['results']['sunset']
    am_pm = datetime.strptime(utc_time, "%I:%M:%S %p")
    mil_time = datetime.strftime(am_pm, "%H:%M")
    sunset_times.append(mil_time)
    print(mil_time)
```

```python
sunset_qual['Sunset Time'] = sunset_times
sunset_qual = sunset_qual[["Date", "City", "Sunset Time", "Good Sunset"]]
sunset_qual
```
